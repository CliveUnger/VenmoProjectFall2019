```python
import json
import pprint
```

```python
# flattens json but preserves arrays as values.
def flattenjson(b, delim):
    val = {}
    for i in b.keys():
        if isinstance(b[i], dict):
            get = flattenjson(b[i], delim)
            for j in get.keys():
                val[i + delim + j] = get[j]
        else:
            val[i] = b[i]

    return val


# flattens json, even flattens the values that are arrays
def flatten_json(y):
    out = {}

    def flatten(x, name=''):
        if type(x) is dict:
            for a in x:
                flatten(x[a], name + a + '_')
        elif type(x) is list:
            i = 0
            for a in x:
                flatten(a, name + str(i) + '_')
                i += 1
        else:
            out[name[:-1]] = x

    flatten(y)
    return out

def pre_proc(entry):
    entry["_id"] = entry["_id"]["$oid"]
    entry['likes'] = len(entry['likes']['data'])
    entry['mentions'] = len(entry['mentions'])
    entry['comments'] = len(entry['comments'])
    entry['target'] = entry['transactions'][0]['target']
    entry.pop("via")
    entry.pop("action_links")
    entry.pop("transactions")

    return entry
```

```python
filename = "/data/06271/cju256/test.json"

with open(filename, 'r') as f:
    entry = json.loads(f.readline())
    entry = pre_proc(entry)
    
    flat = flattenjson(entry,"_")
    pprint.pprint(flat)
    print("\n\n")
```
