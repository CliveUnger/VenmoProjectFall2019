## Degree Distribution

Node degree, the number of edges connected to a node, is one of the most studied properties of networks. Types of networks are often characterized in terms of their distribution of node degrees. We obtain and visualize the degree distribution of our example network as follows.

```python
%matplotlib inline
import matplotlib.pyplot as plt

dd = sorted(nk.centrality.DegreeCentrality(G).run().scores(), reverse=True)
plt.xscale("log")
plt.xlabel("degree")
plt.yscale("log")
plt.ylabel("number of nodes")
plt.plot(dd)
plt.show()
```

## K- Core Decomposition

```python
G.removeSelfLoops()
```

```python
kcore = nk.centrality.CoreDecomposition(G, storeNodeOrder=True)
kcore.run()
```

```python
kcore.maxCoreNumber()
```

```python
kcore.centralization()
```

```python
degrees = {0: []}
for i, s in enumerate(kcore.scores()):
    d = G.degree(i)
    degrees.setdefault(d, []).append(s)
```

```python
import numpy as np

x = []
y = []

x2 = []
y2 = []
for degree in sorted(degrees.keys()):
    if degree < 100:
        x.append(degree)
        y.append(np.mean(degrees[degree]))
    else:
        x2.append(degree)
        y2.append(np.mean(degrees[degree]))
```

```python
import matplotlib.pyplot as plt

plt.xlabel('Node Degree')
plt.ylabel('Avg Core Level')
plt.plot(x, y);
```

```python
plt.yscale("log")
plt.xscale("log")
plt.plot(x2[:-1], y2[:-1])
```

```python
louvainParams = nk.community.PLM(G, \
                                  refine = True, \
                                  maxIter = 5, \
                                  turbo = True, \
                                 )

plmCommunities = nk.community.detectCommunities(G, algo=louvainParams)
```
